```python
import wikipedia
import json
import time

wikipedia.set_lang("en")

characters = [
    "Batman",
    "Iron Man",
    "Sherlock Holmes",
    "Naruto Uzumaki",
    "Joker DC Comics",
    "Thor Marvel",
    "Loki Marvel",
    "Light Yagami",
    "Harry Potter"
]

data = []

for name in characters:
    try:
        print(f"Fetching: {name}")
        
        # Step 1: Search for correct page
        search_results = wikipedia.search(name)

        if not search_results:
            print(f"No results found for {name}")
            continue

        best_match = search_results[0]

        # Step 2: Get page
        page = wikipedia.page(best_match, auto_suggest=False)
        summary = page.summary[:800]  # limit length

        data.append({
            "name": name,
            "wiki_title": page.title,
            "description": summary,
            "url": page.url
        })

        time.sleep(1)

    except Exception as e:
        print(f"Error fetching {name}: {e}")

with open("characters.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

print("Done! Clean dataset saved.")
```

```python
import requests
import json
import time
import urllib.parse

characters = [
    "Batman",
    "Iron Man",
    "Sherlock Holmes",
    "Naruto Uzumaki",
    "Joker (DC Comics)",
    "Thor (Marvel Comics)",
    "Loki (Marvel Comics)",
    "Light Yagami",
    "Harry Potter"
]

headers = {
    "User-Agent": "VectorSearchProject/1.0"
}

data = []

for name in characters:
    try:
        print(f"Fetching: {name}")
        
        encoded_name = urllib.parse.quote(name)
        url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{encoded_name}"
        
        response = requests.get(url, headers=headers)
        print("Status:", response.status_code)

        if response.status_code != 200:
            print(f"Failed for {name}")
            continue
        
        result = response.json()
        
        description = result.get("extract", "")
        image_url = result.get("thumbnail", {}).get("source")

        data.append({
            "name": name,
            "description": description,
            "url": result.get("content_urls", {}).get("desktop", {}).get("page"),
            "image_url": image_url
        })

        time.sleep(1)

    except Exception as e:
        print(f"Error fetching {name}: {e}")

with open("characters.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

print("Done! Clean images saved.")
```
